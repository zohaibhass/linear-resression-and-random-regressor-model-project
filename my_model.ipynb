# **delaney_solubility_with_descriptors**

```python
import pandas as pd
df=pd.read_csv("https://raw.githubusercontent.com/dataprofessor/data/master/delaney_solubility_with_descriptors.csv")
df
```

##separation as x and y

```python
y=df["logS"]
y
```

```python
x=df.drop("logS",axis=1)
x
```

##Data spiliting

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test= train_test_split(x,y,test_size=0.2, random_state=100)
x_train
```

```python
x_test
```

## Building model

### linear regression

**traning the model**

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
lr.fit(x_train, y_train)
LinearRegression()
```

**Applying the model to make the predictions**

```python
y_lr_train_pred = lr.predict(x_train)
y_lr_test_pred=lr.predict(x_test)
```

```python
y_lr_train_pred
```

```python
y_lr_test_pred
```

**Evaluate Model Performance**

```python
from sklearn.metrics import mean_squared_error, r2_score
# Assuming y_lr_train_pred and y_lr_test_pred are methods of some LinearModel object
# You need to call them with input data (x_train and x_test) to obtain the predictions

# For training set
lr_train_pred = y_lr_train_pred(x_train)  # Assuming x_train is the input data
lr_train_mse = mean_squared_error(y_train, lr_train_pred)
lr_train_r2 = r2_score(y_train, lr_train_pred)

# For test set
lr_test_mse = mean_squared_error(y_test, y_lr_test_pred)  # Assuming y_lr_test_pred is the predicted values
lr_test_r2 = r2_score(y_test, y_lr_test_pred)

# Print results
print("Mean squared error (train):", lr_train_mse)
print("R-squared score (train):", lr_train_r2)

print("Mean squared error (test):", lr_test_mse)
print("R-squared score (test):", lr_test_r2)


```

```python
lr_results= pd.DataFrame(['LinearRegression', lr_train_mse, lr_train_r2, lr_test_mse, lr_test_r2]).transpose()
lr_results.columns=['method','Traning ,mse','traning r2','test_mse','test r2']
```

```python
lr_results
```

## **Random Forest**

### **traning the model to make predivtions using randomforest model**

```python
from sklearn.ensemble import RandomForestRegressor
rf=RandomForestRegressor(max_depth=2, random_state=100)
rf.fit(x_train, y_train)
```

### **Evaluate model performance **

```python
y_rf_train_pred = lr.predict(x_train)
y_rf_test_pred=lr.predict(x_test)
```

```python
# Assuming y_rf_train_pred and y_rf_test_pred are methods of some Random Forest model object
# You need to call them with input data (x_train and x_test) to obtain the predictions

# For training set
rf_train_pred = y_rf_train_pred  # Assuming y_rf_train_pred contains the predicted values
rf_train_mse = mean_squared_error(y_train, rf_train_pred)
rf_train_r2 = r2_score(y_train, rf_train_pred)

# For test set
rf_test_mse = mean_squared_error(y_test, y_rf_test_pred)  # Assuming y_rf_test_pred contains the predicted values
rf_test_r2 = r2_score(y_test, y_rf_test_pred)

# Print results
print("Mean squared error (train):", rf_train_mse)
print("R-squared score (train):", rf_train_r2)

print("Mean squared error (test):", rf_test_mse)
print("R-squared score (test):", rf_test_r2)
```

```python
rf_results = pd.DataFrame(['RandomForestRegressor', rf_train_mse, rf_train_r2, rf_test_mse, rf_test_r2]).transpose()
rf_results.columns = ['method', 'Training mse', 'training r2', 'test mse', 'test r2']
rf_results
```

# **models comparisons**

```python
df_models=pd.concat([lr_results,rf_results],axis=0)
df_models
```
